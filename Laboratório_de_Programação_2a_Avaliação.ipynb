# **Quest√£o 1**

```python
# !pip install seaborn
import pandas as pd
import seaborn as sn

dados_filmes = pd.read_csv('https://raw.githubusercontent.com/armandossrecife/lp20231/main/top-500-movies.csv')
dados_filmes

correlation = dados_filmes.corr()
plot = sn.heatmap(correlation, annot = True, fmt=".1f", linewidths=.6)
```

# **Quest√£o 2**

```python
df = dados_filmes.copy()

top_production_cost = df.sort_values(by='production_cost', ascending=False).head(20)
top_worldwide_gross = df.sort_values(by='worldwide_gross', ascending=False).head(20)

print('Top 20 filmes com maior custo de produ√ß√£o:\n')
print(top_production_cost['title'])
print('\n\n\nTop 20 filmes com maior arrecada√ß√£o bilheteria mundial:\n')
print(top_worldwide_gross['title'])
```

# **Quest√£o 3**


```python
generos = ['title', 'genre']

print('G√™neros dos 20 filmes com maior custo de produ√ß√£o:\n')
print(top_production_cost[generos])
print('\n\n\nG√™neros dos 20 filmes com maior arrecada√ß√£o bilheteria mundial:\n')
print(top_worldwide_gross[generos])
```

# **Quest√£o 4**

```python
df = dados_filmes.copy()
top20_worldwide_gross = df.sort_values(by='worldwide_gross', ascending=False).head(20)
media_arrecadacao = top20_worldwide_gross['worldwide_gross'].mean()

print(f'A m√©dia de arrecada√ß√£o mundial dos 20 filmes com maior bilheteria √©: {media_arrecadacao:.2f}')
```
